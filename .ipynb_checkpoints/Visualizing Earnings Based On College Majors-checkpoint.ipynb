# Visualizing Earnings Based On College Majors
## Introduction

In this project we'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Using visualizations, we will explore questions from the dataset like:

- Do students in more popular majors make more money?
    - Using scatter plots
- How many majors are predominantly male? Predominantly female?
    - Using histograms
- Which category of majors have the most students?
    - Using bar plots

#### Data Dictionary
Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are the columns in the dataset:

- `Rank` - Rank by median earnings (the dataset is ordered by this column)
- `Major_code`	 - Major code, FO1DP in ACS PUMS
- `Major` - Major description
- `Major_category` - Category of major from Carnevale et al
- `Total` - Total number of people with major
- `Sample_size` - Sample size (unweighted) of full-time, year-round ONLY (used for earnings)
- `Men` - Male graduates
- `Women` - Female graduates
- `ShareWomen` - Women as share of total
- `Employed` - Number employed (ESR == 1 or 2)
- `Full_time` - Employed 35 hours or more
- `Part_time` - Employed less than 35 hours
- `Full_time_year_round` - Employed at least 50 weeks (WKW == 1) and at least 35 hours (WKHP >= 35)
- `Unemployed` - Number unemployed (ESR == 3)
- `Unemployment_rate` - Unemployed / (Unemployed + Employed)
- `Median` - Median earnings of full-time, year-round workers
- `P25th` - 25th percentile of earnings
- `P75th` - 75th percentile of earnings
- `College_jobs` - Number with job requiring a college degree
- `Non_college_jobs` - Number with job not requiring a college degree
- `Low_wage_jobs` - Number in low-wage service jobs

## Visualizing the Data: Scatter Plots

Using scatter plots we will answer the following questions:

- Do students in more popular majors make more money?
- Is there a link between the number of full-employees and median salary?
- Do students in subjects that were majority female make more money?


#### Employed vs. Sample_size

```python
ax = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
ax.set_title('Employed vs. Sample_size')
```

First it is important that we show that there is a strong positive correlation between the the total number of full-timee eployees with a certain major and the sample size for that major. This is important to keep in mind moving forward as it means that the earnings data for majors with more full-time employees are more representitive of the total population than thos with less.

#### Sample_size vs. Median

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Median vs. Sample_size')
```

**Q. Do students in more popular majors make more money?***

**A. The scatter plot suggests that there is no noticeable relationship between the sample size and the median salary.**

However, based on visuals alone, we can see that there are a few outliers which are distorting the chart size.  This distortion might be hiding a relationship within the bulk of the data.

Let's zoom in to see if there is a relationship within a smaller range of majors that fall within the first 75th percentile of sample sizes.

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_xlim(0, recent_grads['Sample_size'].quantile(.75))
ax.set_title('Median vs. Sample_size up to 75th percentile')
```

The relationship between majors and their median earning in this narrowed selection remains the same as before.

This indicates that there is no strong evidence to show that students in more popular majors make more money.

#### Full_time vs. Median

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Median vs. Full_time')
```

**Q. Is there a link between the number of full-time employees and median salary?**

**A. The scatter plot suggests that there is no noticeable relationship between the number of full time workers and median earnings.**

Once again our plot is distorted by the existence of a few outliers and so we will zoom in inorder to be certain there is no realtionship.

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_xlim(0, recent_grads['Full_time'].quantile(.75))
ax.set_ylim(0, recent_grads['Median'].quantile(.75))
ax.set_title('Median vs. Full_time up to 75th percentile (both axes)')
```

The graph above confirms that there is no correlation between the number of workers and median earnings.

#### ShareWomen vs. Median

```python
ax = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
ax.set_title('Median vs. ShareWomen')
```

**Q. Do students in subjects that were majority female make more money?**

**A. No. The scatter plot suggests that there is negative correlation between the proportion of femal graduates and median earnings.**

This could be due to, amongst others, either or both of the following:

- Women get paid less based on gender-discrimination.
- The types of majors and subsequent carreers that have higher proportions of female graduates have lower pay.

## Visualizing the Data: Histograms

Using histograms we will examine the following:

- What is the most common median salary range?
- What percentage of majors is predominantly female? 

#### Sample_size

```python
ax = recent_grads['Sample_size'].plot(kind='hist')
ax.set_xlabel('Sample_size')
```

As most of the Sample_size values are found below 500, it will be useful to examine the values below 500 for a more dettailed view.

```python
ax = recent_grads['Sample_size'].hist(bins=50, range=(0,500))
ax.set_xlabel('Sample_size')
```

This graph shows us that most sample sizes where below 100, which raises concerns over how representative the earnings data for each majpor is.

#### Median

```python
ax = recent_grads['Median'].hist(bins=20, range=(20000,120000))
ax.set_xlabel('Median')
```

**Q. What is the most common median salary range?**

**A. From the above we can determine that majority of median salaries fall within the range \\$30,000-\\$40,000.**

We see quick drop off in frequency for salary bands on either side of it.

#### ShareWomen

```python
ax = recent_grads['ShareWomen'].hist(bins=2)
ax.set_xlabel('ShareWomen')
```

**Q. What percentage of majors is predominantly female?**

**A. The graph above shows that nearly 60% of all majors are majority female.**
